```python
import os
import pandas as pd
from dotenv import load_dotenv
from pycaret.clustering import *

load_dotenv()

PATH = os.getenv("GENERATED_DATASET_PATH")

df_atv_a_fazer = pd.read_csv(f'{PATH}/df_atv_a_fazer.csv')
```

```python
s = setup(df_atv_a_fazer, session_id=42)
```

#### üìä Aplicando os algoritmos de clustering

##### üíªÔ∏è K-Means Clustering

```python
kmeans = create_model("kmeans", num_clusters=3)
```

```python
evaluate_model(kmeans)
```

```python
assign_model_kmeans = assign_model(kmeans)
assign_model_kmeans
```

```python
# **List of available plots (ID - Name):**

# 1. 'cluster' - Cluster PCA Plot (2d)  
# 2. 'tsne' - Cluster t-SNE (3d)  
# 3. 'elbow' - Elbow Plot  
# 4. 'silhouette' - Silhouette Plot  
# 5. 'distance' - Distance Plot  
# 6. 'distribution' - Distribution Plot  

plot_model(kmeans, plot="cluster")
plot_model(kmeans, plot="tsne")
plot_model(kmeans, plot="elbow")
plot_model(kmeans, plot="silhouette")
plot_model(kmeans, plot="distance")
plot_model(kmeans, plot="distribution")
```
